Praktikum 1

Praktikum 1 ini akan membuat JST sederhana (2 layer) dengan forward pass dan backpropagation manual.

Langkah:

1. Buat dataset sederhana (XOR).

2. Inisialisasi bobot dan bias.

3. Implementasikan forward pass.

4. Hitung error dan lakukan backpropagation.

5. Update bobot menggunakan gradient descent.

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 2
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

# Training
for epoch in range(10000):
    # Forward pass
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Hitung error
    error = y - a2

    # Backpropagation
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss}")

# Output akhir
print("Prediksi:")
print(a2)
```

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 3   # Ubah menjadi 3 neuron
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi aktivasi
def sigmoid(x): return 1 / (1 + np.exp(-x))
def sigmoid_derivative(x): return x * (1 - x)

# Training
for epoch in range(10000):
    # Forward
    z1 = np.dot(X, W1) + b1
    a1 = sigmoid(z1)
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Error
    error = y - a2

    # Backprop
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * sigmoid_derivative(a1)
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update bobot
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {np.mean(np.square(error))}")

print("Prediksi akhir:")
print(a2)
```

b. Menambahkan fungsi ReLU

```python
import numpy as np

# Dataset XOR
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

# Parameter
input_size = 2
hidden_size = 3   # Ubah menjadi 3 neuron
output_size = 1
lr = 0.1

# Inisialisasi bobot
W1 = np.random.randn(input_size, hidden_size)
b1 = np.zeros((1, hidden_size))
W2 = np.random.randn(hidden_size, output_size)
b2 = np.zeros((1, output_size))

# Fungsi ReLU
def relu(x): return np.maximum(0, x)
def relu_derivative(x): return np.where(x > 0, 1, 0)

for epoch in range(10000):
    # Forward
    z1 = np.dot(X, W1) + b1
    a1 = relu(z1)  # ganti sigmoid -> ReLU
    z2 = np.dot(a1, W2) + b2
    a2 = sigmoid(z2)

    # Error
    error = y - a2

    # Backprop
    d_a2 = error * sigmoid_derivative(a2)
    d_W2 = np.dot(a1.T, d_a2)
    d_b2 = np.sum(d_a2, axis=0, keepdims=True)

    d_a1 = np.dot(d_a2, W2.T) * relu_derivative(z1)  # ganti sigmoid_derivative -> relu_derivative
    d_W1 = np.dot(X.T, d_a1)
    d_b1 = np.sum(d_a1, axis=0, keepdims=True)

    # Update
    W1 += lr * d_W1
    b1 += lr * d_b1
    W2 += lr * d_W2
    b2 += lr * d_b2

    if epoch % 1000 == 0:
        print(f"Epoch {epoch}, Loss: {np.mean(np.square(error))}")

print("Prediksi akhir:")
print(a2)
```

Selanjutnya, hidden layer tetap menggunakan 3 neuron, namun fungsi aktivasi pada hidden layer diubah menjadi ReLU. Hasil training ditunjukkan berikut:

| Konfigurasi                | Loss Akhir (Epoch 9000) |
| -------------------------- | ----------------------- |
| Hidden = 2 (awal, sigmoid) | **0.00248953**          |
| Hidden = 3 (sigmoid)       | **0.00394637**          |
| Hidden = 3 + ReLU          | **0.00041069**          |


Dari tabel tersebut diperoleh beberapa kesimpulan penting:

a. Perbandingan dengan Hidden 3 Neuron (Sigmoid)

- Penggunaan ReLU menyebabkan penurunan loss yang sangat signifikan.

- Loss akhir ReLU lebih kecil sekitar 89.6% dibandingkan hidden=3 sigmoid.

- ReLU mempercepat konvergensi dan menghasilkan output prediksi yang lebih mendekati target.

b. Perbandingan dengan Konfigurasi Awal

- Konfigurasi 3 neuron + ReLU menghasilkan loss akhir yang 83.5% lebih rendah dibanding konfigurasi awal.

- Hal ini menunjukkan bahwa ReLU lebih efektif dalam menghindari saturasi gradien yang sering terjadi pada sigmoid.

Kesimpulan Umum

- Penambahan jumlah neuron hidden layer dari 2 menjadi 3 tidak otomatis meningkatkan performa. Pada percobaan ini, loss justru sedikit meningkat.

- Penambahan fungsi aktivasi ReLU pada hidden layer memberikan peningkatan performa yang sangat signifikan.

- ReLU membantu jaringan belajar lebih cepat dan mencapai loss yang lebih rendah dibanding semua konfigurasi sebelumnya.

- Kombinasi hidden = 3 neuron + ReLU merupakan konfigurasi terbaik dari tiga eksperimen yang dilakukan.
